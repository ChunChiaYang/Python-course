### matplotlib.animation.FuncAnimation() :藉由重複畫圖來製造動畫
  
<a href='https://matplotlib.org/api/_as_gen/matplotlib.animation.FuncAnimation.html'><font size='2'>參考資料:<font></a>


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure()
ax = plt.axes(xlim=(-10, 10),ylim=(0,100))
x, y = [], []
line, = plt.plot([], [],'ro',markersize=3)


def init():
    line.set_data([], [])
    return line,

def update(frame):
    x = np.linspace(-10,-10+frame,num=frame+1)
    y= x**2
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=21,
                    init_func=init,interval=200)

#Set anime format
ax.set_aspect(1./ax.get_data_ratio())
plt.title('y=x^2')
plt.ylabel('y',rotation=0)
plt.xlabel('x')
plt.grid()
plt.show()
```

### 總結一下
FuncAnimation()每次在設定好的圖片格式上透過更新函式畫上新的點，達到動畫的效果
- frame:循環長度
- interval:每張圖片的間隔時間

### Practice
請生成25個等分點並透過動畫效果畫出: y=f(x)=2^(-x)+2  
x的範圍請限制在-3到3之間，播放速度可以任意調整


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure()
ax = plt.axes(xlim=(-6,6),ylim=(0,40))
x, y = [], []
line, = plt.plot([], [],'ro',markersize=3)


def init():
    line.set_data([], [])
    return line,

def update(frame):
    """
    To do: Set x and y
    """
    line.set_data(x, y)
    return line,

"TO do:Set frames"
ani = FuncAnimation(fig, update, frames=' ',
                    init_func=init,interval=200)

#Set anime format
ax.set_aspect(1./ax.get_data_ratio())
plt.grid()
plt.title('y=2^(-x)+2')
plt.ylabel('y',rotation=0)
plt.xlabel('x')
plt.show()
```

### Advanced Practice
請試著用動畫效果畫出一個半徑為1的圓  
  
**提示:**  
你可能會用到以下兩個函式:numpy.sin()、numpy.cos()  
在Numpy中，圓周率π可以用:numpy.pi

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
xdata, ydata = [], []
line, = plt.plot([], [], 'ro', animated=True)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    """
    To do: Set x and y
    """
    line.set_data(xdata, ydata)
    return line,

"TO do:Set frames"
ani = FuncAnimation(fig, update, frames=' ',
                    init_func=init, blit=True)
ax.set_aspect(1./ax.get_data_ratio())
plt.show()
```
